```python
#Q2
"""Scatter Plot: Displays data points as individual markers, useful for showing relationships between variables."""

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.scatter(x,y)
plt.show()
```

```python
#Q3
"""The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to organize plots in a grid layout, making it easier to compare and analyze multiple datasets simultaneously."""

import matplotlib.pyplot as plt
import numpy as np

# Data for the lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o')
plt.title('Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='s')
plt.title('Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='^')
plt.title('Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='d')
plt.title('Line 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
#Q4
"""A bar plot (or bar chart) is a type of plot that represents categorical data with rectangular bars, where the length or height of each bar is proportional to the value it represents. Bar plots are useful for comparing different categories or groups of data"""

import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.xlabel('Companies')
plt.ylabel('Profit')
plt.title('Barplot')
plt.bar(company,profit)

plt.show()
```
